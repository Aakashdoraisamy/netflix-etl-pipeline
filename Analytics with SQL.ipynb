```python
%sql
USE CATALOG netflix;
USE SCHEMA etl;
```

# Movies vs TV Shows

## **Insight:**  
- Movies dominate Netflix’s catalog (~70%), while TV Shows make up around 30%.  
- This indicates Netflix prioritizes movies, but TV content is also steadily growing.

```python
%sql
SELECT type, count AS total
FROM gold_type_counts
ORDER BY total DESC;
```

# Titles added per year

## **Insight:**  
- Netflix saw very few additions before 2015.  
- After 2015, there was rapid growth, peaking around 2018–2019.  
- The decline after 2020 may reflect content saturation or shifting strategy.

```python
%sql
SELECT year_added, count AS total
FROM gold_yearly
WHERE year_added IS NOT NULL
ORDER BY year_added;
```

# Top 10 Countries by Titles

## **Insight:**  
- The United States leads by a wide margin, with India and the United Kingdom following.  
- India’s high ranking shows it is Netflix’s second-largest market after the US.  
- Other contributors include Japan, South Korea, and Canada, showing global diversity.

```python
%sql
SELECT country, count AS total
FROM gold_country
WHERE country IS NOT NULL
ORDER BY total DESC
LIMIT 10;
```

# Top 15 Genres

## **Insight:**  
- *International Movies*, *Dramas*, and *Comedies* dominate the catalog.  
- Genres like *Documentaries* and *Action & Adventure* also show strong presence.  
- The wide spread indicates Netflix’s strategy to serve both entertainment and informative content.

```python
%sql
SELECT genre, count AS total
FROM gold_genres
WHERE genre IS NOT NULL
ORDER BY total DESC
LIMIT 15;
```

# Most Common Ratings

## **Insight:**  
- TV-MA is the most frequent rating, followed by TV-14 and PG-13.  
- This shows Netflix targets a **mature and teen audience** primarily.  
- Family/kids ratings (G, PG) exist but represent a much smaller share.

```python
%sql
SELECT rating, COUNT(*) as total
FROM silver_netflix
WHERE rating IS NOT NULL AND rating <> 'Unknown'
GROUP BY rating
ORDER BY total DESC
LIMIT 10;
```

# Movie Duration Distribution

**Insight:**  
- Most Netflix movies cluster between **90–120 minutes**, which is the standard film length.  
- A smaller number of short films (<60 mins) and long features (>150 mins) also exist.  
- This distribution suggests Netflix aligns with traditional movie runtimes.

```python
%sql
SELECT duration_mins
FROM silver_netflix
WHERE type = 'Movie' AND duration_mins IS NOT NULL;
```
